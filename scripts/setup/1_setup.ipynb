#creation of catalog fmcg

```python
%sql

CREATE CATALOG IF NOT EXISTS main_fmcg;
USE CATALOG main_fmcg;
```

#creation of schema s (bronze,silver,gold) 

```python
%sql

CREATE SCHEMA IF NOT EXISTS main_fmcg.gold;
CREATE SCHEMA IF NOT EXISTS main_fmcg.silver;
CREATE SCHEMA IF NOT EXISTS main_fmcg.bronze;
```
